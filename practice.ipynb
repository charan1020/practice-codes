```python
def count_consecutive(text):
    result = []
    count = 1

    for i in range(1, len(text)):
        if text[i] == text[i-1]:
            count += 1
        else:
            result.append(f"{text[i-1]}:{count}")
            count = 1
    result.append(f"{text[-1]}:{count}")  # add last group

    return result
# ---- Usage ----
input_str = "aaaabbbcc"
output = count_consecutive(input_str)
for item in output:
    print(item)
```

```python
from itertools import groupby

input_str = "aaaabbbcc"
result = [f"{char}:{sum(1 for _ in group)}" for char, group in groupby(input_str)]
for item in result:
    print(item)
```

```python
#Armstrong number(153=1^3+5^3+3^3=153)
def is_armstrong(num):
    num_str =str(num)
    n = len(num_str)
    total=0
    for digit in num_str:
        total+= int(digit) ** n
    return total == num
number=int(input("Enter a number:"))
if is_armstrong(number):
    print(f"{number} is an armstrong number")
else:
    print(f"{number} is not an armstrong number")

```

```python
#while loop
num=int(input("Enter a number:"))
original_num=num
total=0
n=len(str(num))
while num>0:
    digit = num % 10
    total+= digit **n
    num//=10 # floor division
if total== original_num:
    print(f"{original_num} is an armstrong number")
else:
    print(f"{origin_num} is not an armstrong number")

    
```

```python
#interval from user of armstrong number
lower=int(input("enter the lower limit of the interval:"))
upper=int(input("enter the upper limit of the interval"))
for num in range(lower,upper+1):
    order=len(str(num))#find the number of digits in num
    original_num=num
    sum=0
    while original_num>0:
        digit = original_num%10
        sum +=digit ** order
        original_num//=10 # in original place we can use 'temp'
    if num==sum:
        print(num,end=" ")
```
