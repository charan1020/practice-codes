```python
#Count even & odd numbers in array
def count_even_odd(arr):
    even = odd = 0
    for i in arr:
        if i % 2 == 0:
            even += 1
        else:
            odd += 1
    return even, odd

arr = [10, 15, 20, 25, 30]
print(count_even_odd(arr))

```

```python
#Linear search-Linear search checks each element one by one until the target value is found or the list ends.
arr = list(map(int, input().split()))
key = int(input())
found = False
for i in range(len(arr)):
    if arr[i] == key:
        print("Element found at index", i)
        found = True
        break
if not found:
    print("Element not found")
```

```python
#Binary search finds an element by repeatedly dividing the search space into half.
arr = list(map(int, input().split()))
key = int(input())
low = 0
high = len(arr) - 1
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print("Element found at index", mid)
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
else:
    print("Element not found")
```

```python
#Armstrong number
num = int(input())
temp = num
n = len(str(num))
s = 0

while temp > 0:
    digit = temp % 10
    s += digit ** n
    temp //= 10

if s == num:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
s = input()
char_set = set()
left = 0
max_len = 0
for right in range(len(s)):
    while s[right] in char_set:
        char_set.remove(s[left])
        left += 1
    char_set.add(s[right])
    max_len = max(max_len, right - left + 1)

print(max_len)
```

```python
#1Ô∏è‚É£ Bubble Sort
arr = list(map(int, input().split()))
n = len(arr)

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print(arr)
```

```python
#2Ô∏è‚É£ Selection Sort
arr = list(map(int, input().split()))
n = len(arr)

for i in range(n):
    min_idx = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_idx]:
            min_idx = j

    arr[i], arr[min_idx] = arr[min_idx], arr[i]

print(arr)
```

```python
#3Ô∏è‚É£ Insertion Sort
arr = list(map(int, input().split()))

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key

print(arr)
```

```python
#Merge two sorted arrays
a = list(map(int, input().split()))
b = list(map(int, input().split()))

i = j = 0
merged = []

while i < len(a) and j < len(b):
    if a[i] <= b[j]:
        merged.append(a[i])
        i += 1
    else:
        merged.append(b[j])
        j += 1

while i < len(a):
    merged.append(a[i])
    i += 1

while j < len(b):
    merged.append(b[j])
    j += 1

print(merged)
```

```python
def reverse(arr, l, r):
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1
        r -= 1

arr = list(map(int, input().split()))
k = int(input())
n = len(arr)

k = k % n

reverse(arr, 0, n-1)
reverse(arr, 0, k-1)
reverse(arr, k, n-1)

print(arr)

```

```python
s = input()
stack = []
pairs = {')': '(', '}': '{', ']': '['}
for ch in s:
    if ch in pairs.values():   # opening brackets
        stack.append(ch)
    elif ch in pairs:          # closing brackets
        if not stack or stack[-1] != pairs[ch]:
            print("Invalid")
            break
        stack.pop()
else:
    if not stack:
        print("Valid")
    else:
        print("Invalid")
```
