# star patterns

```python
num=6
for i in range(num):
    print("*")
```

```python
#right angled triangle(left-aligned)
num=6
for i in range(num):
    print("*"*i)
```

```python
n=10
for i in range(1,n+1):
    print(" *"*i)
```

```python
#right angled triangle(right alined)
num=6
for i in range(num):
    print(" "*(num-i-1),"*"*i)
```

```python
n=10
for i in range(1,n+1):
    print("  "* (n-i)+ "* "* i)
```

```python
#pyramid pattern
num=6
for i in range(num):
    print(" "*(num-i-1),"*"*(2*i+1))
```

```python
n=10
for i in range(1,n+1):
    print(" "* (n-i)+ "* "* i)
```

```python
#pyramid pattern with spaces
num=6
for i in range(num):
    for j in range(num-i-1):
        print(" ",end=" ")
    for j in range(2*i+1):
        print("*",end=" ")
    print()
    
```

```python
#inverted pyramid
n=10
for i in range(n,0,-1):
    print(" "*(n-i)+"* "*i)
```

```python
num = 6
for i in range(num):
    print(" " * i, "*" * (2 * (num - i) - 1))
```

```python
#Diamon pattern
n=10
for i in range(1,n+1):
    print(" "* (n-i)+"* "*i)
for i in range(n-1,0,-1):
    print(" "*(n-i)+"* "*i)
```

```python
#square pattern
n=10
for i in range(1,n+1):
    print("* " *n)
```

# number patterns

```python
#square of numbers
n=10
for i in range(1,n+1):
    for j in range(1,n+1):
        print(j, end=" ")
    print()
```

```python
#triangle of numbers
n=10
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#repated numbers in rows
n=5
for i in range(1,n+1):
    print((str(i)+ " ") *i)
```

```python
#floyds triangle
num=1
n=8
for i in range(1,n+1):
    for j in range(i):
        print(num, end=" ")
        num+=1
    print()
```

```python
#reverse number triangle
n=10
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

# Alphabet patterns

```python
#alphabet triangle
n=5
for i in range(n):
    for j in range(i+1):
        print(chr(65+j),end=" ")
    print()
```

```python
#repeating alphabet triangle
n=5
for i in range(n):
    for j in range(i+1):
        print(chr(65+i),end=" ")
    print()
```

```python
n=5
for i in range(n):
    print(chr(65+i)*(i+1))
```
